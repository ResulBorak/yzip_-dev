```python
import pandas as pd
```

```python
data = pd.read_csv("Chocolate Sales (1).csv")
data
```

## info() fonksiyonu

Bu fonksiyon ilgili data ile ilgili genel bilgileri döndürür

```python
f1 = data.info()
f1
```

## describe() fonksiyonu

Bu fonksiyon datayla ilgili sayısal istatistikleri(mod,medyan,ortalama değer...) döndürür

```python
f2 = data.describe()
f2
```

## astype() fonksiyonu

Bir sütunun veri tipini değiştirmek için kullanılır.

```python
data["Boxes Shipped"] = data["Boxes Shipped"].astype(float)

data["Amount"] = data["Amount"].str.replace("[$, ]", "", regex=True).astype(int)
```

## sort()_values 

Belirli bir sütuna göre veriyi sıralamak için kullanılır.

```python
f4 = data.sort_values(by="Boxes Shipped", ascending=False) #Boxes Shipped sütununa göre azalan sıralama
f4
```

## iloc[] ve loc[]

 iloc[n] fonksiyonu belirtilen satırı getirir, default olarak çalışmadığı için sayı girilmezse hata verir.

 
 iloc[n:m] fonksiyonu bir aralık verilerek de çalıştırılabilir. (n değeriyle başlayıp m-1 değerine sahip veriye kadar tablo döndürür)

```python
data
```

```python
f5 = data.iloc[3]
f5
```

```python
data.iloc[-5:-1] 
```

### loc[] fonkiyonu belirlenen sütunda istenilen koşulu sağlayan verileri döndürür

```python
f5 = data.loc[data["Boxes Shipped"]>100]
f5
```

## nunique() fonksiyonu

Her sütunda kaç farklı değer olduğunu döndürür

```python
f6 = data.nunique()
f6
```

### nlargest() ve nsmallest()

Belirli bir sütundaki en büyük veya en küçük n değerini döndürür.

```python
f7 = data.nlargest(10,"Boxes Shipped")
f7
```

```python
f7 = data.nsmallest(10,"Boxes Shipped")
f7
```

## crosstab() fonksiyonu

İki sütun arasındaki ilişkileri gösteren çapraz tablo oluşturur.

```python
f8 = pd.crosstab(data["Amount"], data["Sales Person"])
f8
```

## isna() ve notna() 

isna(): Herhangi bir verinin kaç tane boş değeri olduğunu döndürür

notna(): Herhangi bir verinin kaç tane dolu değeri olduğunu döndürür

```python
f9 = data.isna().sum()
f9
```

```python
f9 = data.notna().sum()
f9
```

## corr() fonksiyonu

İki veri arasındaki korelasyonu ölçer

Korelasyon ya da ilinti, olasılık kuramı ve istatistikte iki rassal değişken arasındaki doğrusal ilişkinin yönünü ve gücünü belirtir. Genel istatistiksel kullanımda korelasyon, bağımsızlık durumundan ne kadar uzaklaşıldığını gösterir.

```python
f10 = data[["Amount", "Boxes Shipped"]].corr()
f10
```
